```python
from tkinter import *
root=Tk()
root.geometry("500x500")
root.minsize(450,450)
root.maxsize(550,550)

expression = ""

def press(num):
    # point out the global expression variable
    global expression

    # concatenation of string
    expression = expression + str(num)

    # update the expression by using set method
    equation.set(expression)
    
def equalpress():

    try:

        global expression

        total = str(eval(expression))

        equation.set(total)

        expression = ""

    except:

        equation.set(" error ")
        expression = ""

def clear():
    global expression
    expression = ""
    equation.set("")
    

root.title("Simple Calculator")
equation = StringVar()


expression_field = Entry(root, textvariable=equation)
expression_field.grid(columnspan=4, ipadx=70)

b1=Button(root,fg="white",bg="black", text="1",command=lambda:press(1), height=1,width=7)
b1.grid(row=2,column=0)

b2=Button(root,fg="white",bg="black", text="2",command=lambda:press(2), height=1,width=7)
b2.grid(row=2,column=1)

b3=Button(root,fg="white",bg="black", text="3",command=lambda:press(3), height=1,width=7)
b3.grid(row=2,column=2)

b4=Button(root,fg="white",bg="black", text="4",command=lambda:press(4), height=1,width=7)
b4.grid(row=3,column=0)

b5=Button(root,fg="white",bg="black", text="5",command=lambda:press(5), height=1,width=7)
b5.grid(row=3,column=1)

b6=Button(root,fg="white",bg="black", text="6",command=lambda:press(6), height=1,width=7)
b6.grid(row=3,column=2)

b7=Button(root,fg="white",bg="black", text="7",command=lambda:press(7), height=1,width=7)
b7.grid(row=4,column=0)

b8=Button(root,fg="white",bg="black", text="8",command=lambda:press(8), height=1,width=7)
b8.grid(row=4,column=1)

b9=Button(root,fg="white",bg="black", text="9",command=lambda:press(9), height=1,width=7)
b9.grid(row=4,column=2)

b0=Button(root,fg="white",bg="black", text="0",command=lambda:press(0), height=1,width=7)
b0.grid(row=5,column=0)


deci=Button(root,fg="white",bg="black", text=".",command=lambda:press("."), height=1,width=7)
deci.grid(row=5,column=1)

add=Button(root,fg="white",bg="black", text="+",command=lambda:press("+"), height=1,width=7)
add.grid(row=2,column=3)

sub=Button(root,fg="white",bg="black", text="-",command=lambda:press("-"), height=1,width=7)
sub.grid(row=3,column=3)


mul=Button(root,fg="white",bg="black", text="*",command=lambda:press("*"), height=1,width=7)
mul.grid(row=4,column=3)

div=Button(root,fg="white",bg="black", text="/",command=lambda:press("/"), height=1,width=7)
div.grid(row=5,column=3)

equal=Button(root,fg="white",bg="black", text="=",command=equalpress, height=1,width=7)
equal.grid(row=5,column=2)


clear=Button(root,fg="white",bg="black", text="Clear",command=clear, height=1,width=7)
clear.grid(row=6,column=0)

root.mainloop()
```
